# 📋 Planificador de Tareas Inteligente

¡Bienvenida/o a tu propio **Planificador de Tareas Inteligente**!  
Es una aplicación sencilla pero completa para organizar tus tareas del día a día, marcarlas como completadas y ver un resumen semanal.  
Además, está pensada para que aprendas y muestres lo que sabes de HTML, CSS y JavaScript. 🚀✨

---

## ✨ ¿Qué hace esta app?

✅ Agregar tareas con nombre, descripción, prioridad y categoría.  
✅ Mostrar una lista de todas las tareas pendientes.  
✅ Filtrar tareas por categoría.  
✅ Marcar tareas como completadas.  
✅ Ver un resumen con el total de tareas completadas y tiempo ahorrado.  
✅ Guardar las tareas en `localStorage` (tu navegador las recuerda aunque cierres la página).

---

## 🧰 Tecnologías usadas

- 🌐 **HTML5** – Estructura de la app.
- 🎨 **CSS3** – Estilos y diseño bonito.
- ⚡ **JavaScript** – Lógica para agregar, mostrar y marcar tareas.
- 💾 **LocalStorage** – Para guardar datos sin base de datos externa.
